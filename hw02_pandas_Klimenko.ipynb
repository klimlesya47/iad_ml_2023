#### Задание 1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.¶

```python
import pandas as pd
import requests
from zipfile import ZipFile
import io
import numpy as np

names = []
df_info = pd.DataFrame()
df_score  = pd.DataFrame()
index = 0

url = 'https://github.com/hse-ds/iad-intro-ds/raw/master/2023/homeworks/Data.zip'

response = requests.get(url, stream=True, verify=False)
with ZipFile(io.BytesIO(response.content)) as myzip:
  for item in myzip.infolist()[1:]:
    df1 = pd.read_csv(myzip.extract(item.filename), sep=",", index_col=0)
    if index < 10:
      df_info = pd.concat([df_info, df1])
      index += 1
    else:
      df_score = pd.concat([df_score, df1])
      index += 1
df = pd.concat([df_info, df_score], axis = 1)
```

**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
import time
num_group = sorted(df['race/ethnicity'].unique())
num_edu = sorted(df['education'].unique())
```

```python
start_1 = time.time()
sum = 0
count = 0
for i in num_group:
  for it, el in df[df['race/ethnicity']==i]['reading score'].iteritems():
    sum += el;
    count += 1
  print("mean", i, ' = ', sum/count)
  sum = 0
  count = 0
end_1 = time.time()
print(f"время выполнения с помощью циклов = {end_1 - start_1}")
```

```python
start_2 = time.time()
sum = 100
for i in num_edu:
  for it, el in df[df['education']==i]['writing score'].iteritems():
    if el < sum:
      sum = el
  print("min", i, ' = ', sum)
  sum = 100
end_2 = time.time()
print(f"время выполнения с помощью циклов = {end_2 - start_2}")
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
    if average_mark['mean'] >= 90: return 'A'
    elif average_mark['mean'] < 90 and average_mark['mean'] >= 80: return 'B'
    elif average_mark['mean'] < 80 and average_mark['mean'] >= 70: return 'C'
    elif average_mark['mean'] < 70 and average_mark['mean'] >= 60: return 'D'
    elif average_mark['mean'] < 60 and average_mark['mean'] >= 50: return 'E'
    return 'F'
df['mean'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
df['Grade'] = df.apply(GetGrade, axis=1)
df.drop('mean', axis= 1 , inplace= True )
df
```
